```python
import string
from collections import Counter

def preprocess_text(text):
    """
    Preprocess the text by converting to lowercase, removing specific symbols,
    and replacing punctuation with spaces.
    """
    text = text.lower()
    
    # Remove the £ symbol
    text = text.replace('£', '')
    
    # Translate punctuation to spaces
    translator = str.maketrans(string.punctuation, ' ' * len(string.punctuation))
    text = text.translate(translator)
    
    # Replace multiple spaces with a single space
    text = ' '.join(text.split())
    
    return text

def count_word_occurrences(text):
    """
    Count occurrences of each word in the text.
    """
    words = text.split()
    return Counter(words)

def calculate_accuracy(reference_text, extracted_text):
    """
    Calculate the accuracy by comparing the reference text with the extracted text.
    Returns accuracy and a list of missing words from the reference text, including incorrect occurrences.
    """
    ref_text = preprocess_text(reference_text)
    ext_text = preprocess_text(extracted_text)
    
    ref_word_counts = count_word_occurrences(ref_text)
    ext_word_counts = count_word_occurrences(ext_text)
    
    # Calculate accuracy based on matching words
    matched_words = set(ref_word_counts.keys()) & set(ext_word_counts.keys())
    accuracy = (sum(min(ref_word_counts[word], ext_word_counts.get(word, 0)) for word in matched_words) / sum(ref_word_counts.values())) * 100 if ref_word_counts else 0
    
    # Find missing words and incorrect occurrences
    missing_words = []
    for word in ref_word_counts:
        ref_count = ref_word_counts[word]
        ext_count = ext_word_counts.get(word, 0)
        
        if ext_count < ref_count:
            missing_count = ref_count - ext_count
            missing_words.extend([word] * missing_count)
    
    return accuracy, missing_words

def compare_texts(reference_file, extracted_file):
    """
    Compare the extracted text with the reference text and determine extraction accuracy.
    """
    with open(reference_file, 'r', encoding='utf-8') as file:
        reference_text = file.read()
    
    with open(extracted_file, 'r', encoding='utf-8') as file:
        extracted_text = file.read()
    
    accuracy, missing_words = calculate_accuracy(reference_text, extracted_text)
    
    print(f"Extraction Accuracy: {accuracy:.2f}%")
    
    if missing_words:
        print("Missing words from the reference text:")
        for word in missing_words:
            print(word)
    else:
        print("No missing words found.")

```

```python
import cv2
import numpy as np
import pytesseract
import time
from wand.image import Image

# Start timing
start_time = time.time()

# Set the path to the Tesseract-OCR executable
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'


output_image_path = r"C:\Users\aelg\Documents\t1res_without_underlines.jpeg"
extracted_text_file = r"C:\Users\aelg\Documents\extracted_text.txt"

# Deskew the image using Wand
with Image(filename=input_image_path) as img:
    img.deskew(0.4 * img.quantum_range)
    img.save(filename=r"C:\Users\aelg\Documents\t1resizeddeskewed.jpeg")
    
# Load the deskewed image
img = cv2.imread(r"C:\Users\aelg\Documents\t1resizeddeskewed.jpeg")
if img is None:
    raise ValueError("Image not loaded. Check the file path.")

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Thresholding
thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

# Invert the image to make the text white and background black
thresh = 255 - thresh

# Apply horizontal morphology close to merge text into lines
kernel = np.ones((5, 191), np.uint8)
morph = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)

# Find external contours for larger text blocks (paragraphs)
contours = cv2.findContours(morph, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contours = contours[0] if len(contours) == 2 else contours[1]

# Initialize a list to store the bounding boxes
bounding_boxes = []

# Set the size threshold for filtering small boxes
min_width = 70
min_height = 20

# Filter and store bounding boxes
for cntr in contours:
    x, y, w, h = cv2.boundingRect(cntr)
    if w >= min_width and h >= min_height:
        bounding_boxes.append((x, y, w, h))

# Merge overlapping or close bounding boxes into larger ones
def merge_boxes(bounding_boxes, pad=20):
    merged_boxes = []
    for box in bounding_boxes:
        x, y, w, h = box
        x -= pad
        y -= pad
        w += 2 * pad
        h += 2 * pad
        
        if not merged_boxes:
            merged_boxes.append([x, y, x + w, y + h])
        else:
            merged = False
            for i in range(len(merged_boxes)):
                mx1, my1, mx2, my2 = merged_boxes[i]
                if x < mx2 and x + w > mx1 and y < my2 and y + h > my1:
                    merged_boxes[i] = [min(mx1, x), min(my1, y), max(mx2, x + w), max(my2, y + h)]
                    merged = True
                    break
            if not merged:
                merged_boxes.append([x, y, x + w, y + h])
    return merged_boxes

# Merge the boxes
merged_boxes = merge_boxes(bounding_boxes)

# Create a copy of the original image to draw bounding boxes and underlines
result = img.copy()

# Open the text file in write mode to clear any previous content
with open(extracted_text_file, 'w', encoding='utf-8') as f:
    pass

# Process each detected paragraph separately
for i, (x1, y1, x2, y2) in enumerate(merged_boxes):
    x1 = max(0, x1)
    y1 = max(0, y1)
    x2 = min(img.shape[1], x2)
    y2 = min(img.shape[0], y2)

    if x1 >= x2 or y1 >= y2:
        print(f"Invalid bounding box: ({x1}, {y1}), ({x2}, {y2})")
        continue

    # Crop the paragraph from the image
    paragraph_img = img[y1:y2, x1:x2]

    # Apply Tesseract OCR to the cropped paragraph
    config = "--psm 6 -l deu"
    text = pytesseract.image_to_string(paragraph_img, config=config)
    
    # Save the extracted text to the file
    with open(extracted_text_file, 'a', encoding='utf-8') as text_file:
        text_file.write(f"Paragraph {i+1}:\n{text}\n\n")
    
    # Print the extracted text
    print(f"Text from paragraph {i+1}:")
    print(text)
    print("\n" + "="*50 + "\n")
    
    # Draw the bounding box on the original image
    cv2.rectangle(result, (x1, y1), (x2, y2), (0, 0, 255), 2)

# Save the final image with bounding boxes
cv2.imwrite(output_image_path, result)

# End timing
end_time = time.time()

# Compare the texts and print the accuracy and missing words
compare_texts(reference_text_file, extracted_text_file)

# Calculate and print the runtime
runtime = end_time - start_time
print(f"Runtime: {runtime:.2f} seconds")
```

```python
import cv2
import numpy as np
import pytesseract
import os
from wand.image import Image
from wand.display import display
# Start timing
start_time = time.time()

# Set the path to the Tesseract-OCR executable
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Function to clear the text file at the beginning of the script
def clear_text_file(filename=r"C:\Users\aelg\Documents\extracted_text.txt"):
    with open(filename, "w", encoding="utf-8") as file:
        pass  # Opening the file in write mode clears its contents

# Call the function to clear the text file
clear_text_file()

# Define the paths
output_image_path = r"C:\Users\aelg\Documents\t1res_with_underlines.jpeg"
temp_dir = r"C:\Users\aelg\Documents\temp_paragraphs"
extracted_text_file = r"C:\Users\aelg\Documents\extracted_text.txt"

# Create the temporary directory if it doesn't exist
os.makedirs(temp_dir, exist_ok=True)

# Deskew the image using Wand
with Image(filename=input_image_path) as img:
    img.deskew(0.4 * img.quantum_range)
    img.save(filename=r"C:\Users\aelg\Documents\t1resizeddeskewed.jpeg")
    
# Load the deskewed image
img = cv2.imread(r"C:\Users\aelg\Documents\t1resizeddeskewed.jpeg")

# Check if the image is loaded correctly
if img is None:
    raise ValueError("Image not loaded. Check the file path.")

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Thresholding
thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

# Invert the image to make the text white and background black
thresh = 255 - thresh

# Apply horizontal morphology close to merge text into lines
kernel = np.ones((5, 191), np.uint8)
morph = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)

# Find external contours for larger text blocks (paragraphs)
contours = cv2.findContours(morph, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contours = contours[0] if len(contours) == 2 else contours[1]

# Initialize a list to store the bounding boxes
bounding_boxes = []

# Set the size threshold for filtering small boxes
min_width = 70
min_height = 20

# Filter and store bounding boxes
for cntr in contours:
    x, y, w, h = cv2.boundingRect(cntr)
    
    # Filter out small boxes based on width and height
    if w >= min_width and h >= min_height:
        bounding_boxes.append((x, y, w, h))

# Merge overlapping or close bounding boxes into larger ones
def merge_boxes(bounding_boxes, pad=20):
    merged_boxes = []
    for box in bounding_boxes:
        x, y, w, h = box
        x -= pad
        y -= pad
        w += 2 * pad
        h += 2 * pad
        
        if not merged_boxes:
            merged_boxes.append([x, y, x + w, y + h])
        else:
            merged = False
            for i in range(len(merged_boxes)):
                mx1, my1, mx2, my2 = merged_boxes[i]
                if x < mx2 and x + w > mx1 and y < my2 and y + h > my1:
                    merged_boxes[i] = [min(mx1, x), min(my1, y), max(mx2, x + w), max(my2, y + h)]
                    merged = True
                    break
            if not merged:
                merged_boxes.append([x, y, x + w, y + h])
    return merged_boxes

# Merge the boxes
merged_boxes = merge_boxes(bounding_boxes)

# Create a copy of the original image to draw bounding boxes and underlines
result = img.copy()

# Define minimum length for underlines
min_line_length = 10

# Open the text file to store the extracted text
with open(extracted_text_file, 'w', encoding='utf-8') as text_file:
    # Process each detected paragraph separately
    for i, (x1, y1, x2, y2) in enumerate(merged_boxes):
        # Ensure coordinates are within bounds
        x1 = max(0, x1)
        y1 = max(0, y1)
        x2 = min(img.shape[1], x2)
        y2 = min(img.shape[0], y2)

        if x1 >= x2 or y1 >= y2:
            print(f"Invalid bounding box: ({x1}, {y1}), ({x2}, {y2})")
            continue

        # Crop the paragraph from the image
        paragraph_img = img[y1:y2, x1:x2]
        cropped_thresh = thresh[y1:y2, x1:x2]

        if cropped_thresh.size == 0:
            print(f"Empty cropped image for coordinates: ({x1}, {y1}), ({x2}, {y2})")
            continue

        # Underline words on the paragraph image itself
        word_kernel = np.ones((3, 20), np.uint8)
        word_morph = cv2.morphologyEx(cropped_thresh, cv2.MORPH_CLOSE, word_kernel)
        word_contours = cv2.findContours(word_morph, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        word_contours = word_contours[0] if len(word_contours) == 2 else word_contours[1]

        # Create an image with underlined words directly on the paragraph image
        for word_cntr in word_contours:
            wx, wy, ww, wh = cv2.boundingRect(word_cntr)
            # Draw underline if the width of the bounding box is above the threshold
            if ww >= min_line_length:
                cv2.line(paragraph_img, (wx, wy + wh + 2), (wx + ww, wy + wh + 2), (0, 0, 128), 2)

        # Resize and enhance the paragraph image
        #scale_factor = 1.17 # or an appropriate factor
        #paragraph_img = cv2.resize(paragraph_img, None, fx=scale_factor, fy=scale_factor, interpolation=cv2.INTER_CUBIC)
        # alpha = 0.75  # Contrast control (1.0-3.0)
        # beta = 5  # Brightness control (0-100)
        # paragraph_img = cv2.convertScaleAbs(paragraph_img, alpha=alpha, beta=beta)

        # Save the underlined paragraph to a temporary file
        temp_paragraph_path = os.path.join(temp_dir, f"paragraph_{i}.png")
        cv2.imwrite(temp_paragraph_path, paragraph_img)

        # Apply Tesseract OCR to the underlined paragraph
        text = pytesseract.image_to_string(paragraph_img, config="--psm 6 -l deu")
        
        # Write the text from the paragraph to the text file
        text_file.write(f"Paragraph {i+1}:\n{text}\n")
        
        # Optional: print or store the text from the paragraph
        print(f"Text from paragraph {i+1}:")
        print(text)
        print("\n" + "="*50 + "\n")
        
        # Draw the bounding box on the original image
        cv2.rectangle(result, (x1, y1), (x2, y2), (0, 0, 255), 2)

# Save the final image with bounding boxes and underlines
cv2.imwrite(output_image_path, result)

# Optional: Cleanup temporary files
# import shutil
# shutil.rmtree(temp_dir)

# Display the final result with the original quality (optional)
# cv2.imshow("Final Result", result)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# End timing
end_time = time.time()

# Compare the texts and print the accuracy and missing words
compare_texts(reference_text_file, extracted_text_file)
# Calculate and print the runtime
runtime = end_time - start_time
print(f"Runtime: {runtime:.2f} seconds")
```
